## Logistic Function

The **logistic function**, also known as the **sigmoid function**, is commonly used to model growth that starts exponentially but slows as it approaches a maximum limit.

The general form of the logistic function is:

$$
f(x) = \frac{L}{1 + e^{-k(x - x_0)}}
$$

Where:
- $L$ is the **carrying capacity** (maximum value the function approaches),
- $k$ is the **growth rate** (steepness of the curve),
- $x_0$ is the **midpoint** (the x-value at which the function reaches half of L ).

### Applications

- **Biology**: Population growth models
- **Economics**: Market saturation
- **Machine Learning**: Activation functions in neural networks
- **Statistics**: Logistic regression

```python
x = np.linspace(-10, 10, 400)

def logistic(x, L,B0,B1):
    z=B0+B1*x
    return L / (1 + np.exp(-z))
b1 = [0.5,1,2,1,1,-1,-2]
b0=   [0,0,0,-2,2,0,0]
logistic_values =[]


n = len(b1)
cols = 3
rows = int(np.ceil(n / cols))

plt.figure(figsize=(5*cols, 4*rows))
for i in range(len(b1)):
    B0, B1 = b0[i], b1[i]
    y = logistic(x, 1, B0, B1)
    if B1 != 0:
        x0 = -B0 / B1
    else:
        x0 = None
    
    # plot in subplot
    ax = plt.subplot(rows, cols, i+1)
    ax.plot(x, y, label="logistic")
    
    # add decision boundary
    if x0 is not None:
        ax.axvline(x=x0, color="red", linestyle="--", label="Decision boundary")
    
    ax.set_ylim(-0.1, 1.1)
    ax.set_title(f"B0={B0}, B1={B1}")
    ax.grid(True)
    ax.legend()
plt.tight_layout()
plt.show()
#When B1 is negative, the sigmoid curve poinst the other direction.
```

We will generate a synthetic dataset using `make_classification` and train a logistic regression model. Then, we will visualize the decision boundary and evaluate the model's accuracy.

**General procedure**

1. Generate the dataset
    - sklearn.datasets.make_classification : https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html
2. Plot the dataset
3. Train a Logistic Regression model
4. Make predictions
5. Evaluate the accuracy
    - *Hint :* https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score


```python
# Solution: 
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np


# Generate synthetic dataset
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("2D Classification Dataset")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, stratify=y)
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

Logistic regression assumes a linear decision boundary. Let's explore some exotic scenarios:
- Overlapping classes


```python
# Overlapping classes
X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, class_sep=0.5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, stratify=y)
model = LogisticRegression(solver="saga",max_iter=1000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_overlap = accuracy_score(y_test, y_pred)
print(accuracy)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("2D Classification Dataset")
plt.show()
```

- Outliers

```python
# Outliers
X, y = make_classification(n_samples=198, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, stratify=y)
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_outlier = accuracy_score(y_test, y_pred)
accuracy_outlier
```

- Non-Linear Boundary
    - https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html#sklearn.datasets.make_moons

```python
# Non-linear boundary
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, stratify=y)
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_moon = accuracy_score(y_test, y_pred)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=50)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("2D Classification Dataset")
plt.show()
```
