```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import ( ModelSpec as MS ,summarize )
import seaborn as sns
from ISLP import confusion_table
from ISLP.models import contrast
from sklearn. discriminant_analysis import \
( LinearDiscriminantAnalysis as LDA ,
QuadraticDiscriminantAnalysis as QDA)
from sklearn. naive_bayes import GaussianNB
from sklearn. neighbors import KNeighborsClassifier
from sklearn. preprocessing import StandardScaler
from sklearn. model_selection import train_test_split
from sklearn. linear_model import LogisticRegression
```

```python
#Loading Smarket data
Smarket= load_data("Smarket")
Smarket_noDir = Smarket.drop("Direction",axis=1)
Smarket
```

```python
Smarket_noDir.corr()
```

```python
Smarket.plot(y='Volume')
```

```python
#We want to use logistic regression to predict up and down.
allvars = Smarket.columns.drop(['Today', 'Direction', 'Year'])
design = MS(allvars) #design matrix
X = design.fit_transform (Smarket)
y = Smarket.Direction == 'Up'

glm = sm.GLM(y,X,
family=sm.families.Binomial ())
results = glm.fit()
summarize(results)
```

```python
results.params #Just coefficients
results.pvalues # Just p-values
```

```python
probs = results.predict () # predict probabilities.
probs[:10]
```

```python
labels = np.array(["Down"] * len(probs))   # match length
labels[probs > 0.5] = "Up"

# Ensure consistent dtype and strip spaces just in case
labels = pd.Series(labels).str.strip()
truth = Smarket.Direction.str.strip()
```

```python
confusion_table(labels , Smarket.Direction )
```
