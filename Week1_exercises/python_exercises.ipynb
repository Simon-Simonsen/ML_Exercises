## Create a virtual environment for the machine learning course

### Why virtual environments 
Helps avoid dependency issues between packages. You are using a package manager such as pip or conda, but ocassionally you can run into the situation of no possible combination of package versions being compatible. Additionally, sometimes errors can later occur despite the package manager initially allowing the installation, which are very hard to debug and fix.   
     
### Don't do everything in base
Yes, you can go on a long time just installing everything in base without any issues. But with so many different packages, problems are bound to occur sooner or later. Additionally, virtual environments help keep the installations for your different project organised.

### ‚ö†Ô∏è A warning about copy vs reference in pandas
When you create a dataframe based on another dataframe, such as by filtering which rows to select or by only selecting some columns, it is a **reference** to the original dataframe. This means that any changes you make to one dataframe also apply to the other.     
If you intend to modify the dataframe you are creating, it is the best if you specify **.copy()** to ensure that it is an actual copy of the original dataframe and not a reference, so any changes you make do not also apply to the original dataframe.
